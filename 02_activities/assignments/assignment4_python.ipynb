```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = "C:/Users/prsrivastava/Downloads/2024 Cultural Festivals Funding Program recipients.json"
data = pd.read_json(file_path)
```

```python
# Data overview
print(data.head())
print(data.info())
```

```python
# Summarize total grant amounts by funding stream
grant_summary = data.groupby("Stream")["Grant"].sum().sort_values(ascending=False)

# Plot total grant amounts by funding stream
plt.figure(figsize=(10, 6))
sns.barplot(x=grant_summary.values, y=grant_summary.index, palette="viridis")
plt.title("Total Grant Amounts by Funding Stream (2024)", fontsize=14, weight="bold")
plt.xlabel("Total Grant Amount (CAD)", fontsize=12)
plt.ylabel("Funding Stream", fontsize=12)
plt.tight_layout()
```

```python
# Save cleaned dataset to CSV for use in Tableau or Excel
cleaned_file_path = "C:/Users/prsrivastava/DSI_Demo/visualization/02_activities/assignments/cleaned_festival_data_2024.csv"
data.to_csv(cleaned_file_path, index=False)
print(f"Cleaned dataset saved to: {cleaned_file_path}")
```
